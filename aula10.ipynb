Aproveitar que já usamos a base IRIS para fixar o conhecimento em KNN, fazer um mini projeto com esse dataset, para fixar o conhecimento em clusterização.

**Objetivo do Projeto**

O objetivo deste mini projeto é aplicar a clusterização para agrupar as amostras de flores Iris em clusters com base em suas características físicas, sem usar a informação da espécie, e depois avaliar como esses clusters se comparam às classificações de espécies reais.

```python
# 1. Importar Bibliotecas Necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# 2. Carregar o Conjunto de Dados
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# 3. Exploração e Visualização dos Dados (Exemplo Simples)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Comprimento da Sépala')
plt.ylabel('Largura da Sépala')
plt.title('Visualização do Conjunto de Dados Iris')
plt.show()
```

```python
# 4. Pré-processamento dos Dados
# Neste caso, o conjunto de dados Iris já está pronto para ser usado.
```

```python
# 5. Aplicação do K-Means para Clusterização
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
# 6. Avaliação dos Clusters
print("Matriz de Confusão:")
print(confusion_matrix(y, kmeans.labels_))
print("\nRelatório de Classificação:")
print(classification_report(y, kmeans.labels_))
```

```python
# 7. Visualização dos Clusters
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('Comprimento da Sépala')
plt.ylabel('Largura da Sépala')
plt.title('Visualização dos Clusters')
plt.show()
```

### **O Que é Clustering?**
Clustering é o processo de dividir um conjunto de dados em grupos, ou "clusters", de modo que os pontos de dados dentro de um mesmo grupo sejam mais semelhantes entre si do que com os de outros grupos.

### **Quando Usamos Clustering?**
O clustering é usado em diversas aplicações, como segmentação de mercado, organização de bibliotecas digitais, análise de redes sociais, e detecção de padrões em dados de saúde ou genéticos.
